```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

date_range = pd.date_range(start="2016-03-22", periods=20)

ks_test_cntl = pd.read_csv("ks.test.csv")
ks_test_cntl = ks_test_cntl.drop(ks_test_cntl.columns[0], axis=1)
threshold = pd.read_csv("threshold.csv")
threshold = threshold.drop(threshold.columns[0], axis=1)

ks_test_cntl['Date'] = date_range
threshold['Date'] = date_range

ks_test_cntl = ks_test_cntl.set_index('Date')
threshold = threshold.set_index('Date')

fig = plt.figure()
fig.set_size_inches(17, 12)

ax1 = plt.subplot2grid(shape=(2,6), loc=(0,0), colspan=2)
ax2 = plt.subplot2grid((2,6), (0,2), colspan=2)
ax3 = plt.subplot2grid((2,6), (0,4), colspan=2)
ax4 = plt.subplot2grid((2,6), (1,1), colspan=2)
ax5 = plt.subplot2grid((2,6), (1,3), colspan=2)

threshold[['V1']].plot(ax=ax1, style='ro-', legend=False)
ks_test_cntl[['V1']].plot(ax=ax1, style='co-', legend=False)
ax1.set_ylim(0, 1)
ax1.set_xlim(pd.Timestamp('2016-03-22'), pd.Timestamp('2016-04-10'))
ax1.set_xlabel("")
ax1.set_ylabel("K-S distance")
ax1.set_title("Dataset 1")

threshold[['V2']].plot(ax=ax2, style='ro-', legend=False)
ks_test_cntl[['V2']].plot(ax=ax2, style='co-', legend=False)
ax2.set_ylim(0, 1)
ax2.set_xlim(pd.Timestamp('2016-03-22'), pd.Timestamp('2016-04-10'))
ax2.set_xlabel("")
ax2.set_ylabel("K-S distance")
ax2.set_title("Dataset 2")

threshold[['V3']].plot(ax=ax3, style='ro-', legend=False)
ks_test_cntl[['V3']].plot(ax=ax3, style='co-', legend=False)
ax3.set_ylim(0, 1)
ax3.set_xlim(pd.Timestamp('2016-03-22'), pd.Timestamp('2016-04-10'))
ax3.set_xlabel("")
ax3.set_ylabel("K-S distance")
ax3.set_title("Dataset 3")

threshold[['V4']].plot(ax=ax4, style='ro-', legend=False)
ks_test_cntl[['V4']].plot(ax=ax4, style='co-', legend=False)
ax4.set_ylim(0, 1)
ax4.set_xlim(pd.Timestamp('2016-03-22'), pd.Timestamp('2016-04-10'))
ax4.set_xlabel("")
ax4.set_ylabel("K-S distance")
ax4.set_title("Dataset 4")

threshold[['V5']].plot(ax=ax5, style='ro-', legend=False)
ks_test_cntl[['V5']].plot(ax=ax5, style='co-', legend=False)
ax5.set_ylim(0, 1)
ax5.set_xlim(pd.Timestamp('2016-03-22'), pd.Timestamp('2016-04-10'))
ax5.set_xlabel("")
ax5.set_ylabel("K-S distance")
ax5.set_title("Dataset 5")

plt.tight_layout()
plt.subplots_adjust(top=0.8)

plt.savefig('ks_distance.png')
```
